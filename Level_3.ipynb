# Level_3 – Task_2: Votes Analysis

## Objectives:

Identify the restaurants with the highest and lowest number of votes.
Analyze whether the number of votes affects the aggregate rating.
Visualize the relationship between votes and ratings.

### 1. Top 10 Restaurants with Highest Votes

```python
top_votes = df.sort_values(by="Votes", ascending=False)
top_votes[["Restaurant Name", "Votes", "Aggregate rating"]].head(10)
```

### 2. Restaurants with Lowest Votes

```python
low_votes = df.sort_values(by="Votes", ascending=True)
low_votes[["Restaurant Name", "Votes", "Aggregate rating"]].head(10)
```

### 3. Visualize Top 10 Restaurants by Votes

```python
plt.figure(figsize=(10,5))
top_votes.head(10).plot(x="Restaurant Name",y="Votes", kind="bar",color="orange",legend=False)
plt.title("Top 10 Restaurants by Votes")
plt.xlabel("Restaurant Name")
plt.ylabel("Votes")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

### 4. Relationship Between Votes and Ratings

```python
plt.figure(figsize=(8,6))
plt.scatter(df["Votes"], df["Aggregate rating"], alpha=0.5, color="blue")
plt.title("Votes vs Aggregate Rating")
plt.xlabel("Votes")
plt.ylabel("Aggregate Rating")
plt.show()
```

### 5. Correlation Between Votes and Ratings

```python
correlation = df["Votes"].corr(df["Aggregate rating"])
print("Correlation between Votes and Aggregate Rating:", correlation)
```

## Conclusion

The analysis shows that restaurants with the highest votes generally have good aggregate ratings, while restaurants with very few or zero votes often have low or no ratings. The scatter plot indicates a positive relationship between votes and ratings. The correlation coefficient of approximately **0.31** suggests a weak to moderate positive correlation, meaning that restaurants with more customer votes tend to have higher ratings, although votes alone do not fully determine a restaurant's rating.

# Level_3 – Task_3: Price Range vs Online Delivery & Table Booking

## Objective

Analyze how the restaurant **Price Range** is related to **Online Delivery** and **Table Booking** services. Also visualize the distribution of these services across different price ranges.

### 1. Price Range Distribution

```python
price_range = df["Price range"].value_counts().sort_index()
print(price_range)
```

### 2. Visualize Price Range Distribution

```python
plt.figure(figsize=(6,4))
price_range.plot(kind="bar", color="skyblue")
plt.title("Restaurants by Price Range")
plt.xlabel("Price Range")
plt.ylabel("Number of Restaurants")
plt.show()
```

### 3. Online Delivery vs Price Range

```python
online_delivery = pd.crosstab(df["Price range"],df["Has Online delivery"])
print(online_delivery)
```

### 4. Visualize Online Delivery

```python
online_delivery.plot(kind="bar",figsize=(8,5))
plt.title("Price Range vs Online Delivery")
plt.xlabel("Price Range")
plt.ylabel("Number of Restaurants")
plt.xticks(rotation=0)
plt.show()
```

### 5. Table Booking vs Price Range

```python
table_booking = pd.crosstab(df["Price range"],df["Has Table booking"])
print(table_booking)
```

### 6. Visualize Table Booking

```python
table_booking.plot(kind="bar",figsize=(8,5))
plt.title("Price Range vs Table Booking")
plt.xlabel("Price Range")
plt.ylabel("Number of Restaurants")
plt.xticks(rotation=0)
plt.show()
```

### 7. Average Rating by Price Range

```python
rating_price = df.groupby("Price range")["Aggregate rating"].mean()
print(rating_price)
```

### 8. Visualize Average Rating

```python
plt.figure(figsize=(6,4))
rating_price.plot(kind="bar",color="green")
plt.title("Average Rating by Price Range")
plt.xlabel("Price Range")
plt.ylabel("Average Rating")
plt.xticks(rotation=0)
plt.show()
```

## Conclusion

- Most restaurants belong to **Price Range 1**, while **Price Range 4** has the fewest restaurants.
- Online delivery is most commonly available in **Price Range 2** restaurants, whereas higher price range restaurants are less likely to offer online delivery.
- Table booking is rarely available in **Price Range 1** restaurants but becomes increasingly common in **Price Range 3** and **Price Range 4** restaurants.
- The average aggregate rating increases as the price range increases. Price Range 4 restaurants have the highest average rating, while Price Range 1 restaurants have the lowest.
- Overall, higher-priced restaurants tend to provide premium services such as table booking and also receive better customer ratings.
