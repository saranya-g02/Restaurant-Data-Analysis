# Level_2 - Task-1 : Restaurant Ratings

## Objective

The objective of the task is to analyze the distribution of aggrregate ratings, identify the most common rating range, and calculate the average no.of votes received by restaurants.

### 1. Import Required libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### 2. Load the dataset

```python
df = pd.read_csv("../Dataset/Cognifyz Dataset.csv")
```

### 3. Display the first 5 records

```python
df.head()
```

### 4. Dataset information

```python
df.shape
```

```python
df.columns
```

### 5. Analyze the distribution of aggregate ratings and determine the most common rating range

```python
rating_distribution = df["Aggregate rating"].value_counts().sort_index()
rating_distribution
```

### 6. Visualization of the distribution of Aggregate Ratiings

```python
plt.figure(figsize=(8,5))
rating_distribution.plot(kind="bar", color="purple")
plt.title("Distribution of Aggregate Ratings")
plt.xlabel("Aggregate Ratings")
plt.ylabel("No.of Restaurants")
plt.show()
```

### 7. Find the most common rating range

```python
most_common_rating = rating_distribution.idxmax()
highest_count = rating_distribution.max()
print("Most Common Rating:", most_common_rating)
print("No.of Restaurants:", highest_count)
```

### 8. Calculate the average no.of votes received by restaurants

```python
average_votes = df["Votes"].mean()
print("Average No.of Votes:", round(average_votes,2))
```

### Conclusion

1. The most common rating in the dataset is 0.0, with 2148 restaurants.
2. Most restaurants have ratings between 3.0 and 4.0, as shown in the distribution chart
3. The average no.of votes received by restaurants is 156.91.
4. Higher-rated restaurants generally receive more customer engagement through votes
5. This analysis helps understand the overall rating distribution and customer voting behavior.

# Level_2 - Task_2 :Cuisine Combination

## Objective

The objective of this task is to identify the most common combinations of cuisines served by restaurants and analyze the popularity of different cuisine combinations in the dataset.


### 1. Identify the most common cuisine combinations

```python
cuisine_counts = df["Cuisines"].value_counts()
cuisine_counts.head(10)
```

### 2. Find the most common cuisine combination

```python
most_common_cuisine = cuisine_counts.idxmax()
highest_count = cuisine_counts.max()
print("Most common cuisine combinations:", most_common_cuisine)
print("No.of Restaurants:", highest_count)
```

### 3. Visualize the Top 10 Cuisine Combinations

```python
plt.figure(figsize=(8,5))
cuisine_counts.head(10).plot(kind="bar",color="Orange")
plt.title("Top 10 Most Common Cuisine Combinations")
plt.xlabel("Cuisine Combinations")
plt.ylabel("No.of Restaurants")
plt.xticks(rotation =35, ha="right")
plt.show()
```

### 4. Average rating for each cuisine combination

```python
cuisine_rating = df.groupby("Cuisines")["Aggregate rating"].mean().sort_values(ascending=False)
cuisine_rating.head(10)
```

### 5. Visulize Top 10 Highest Rated Cuisine Combinations

```python
plt.figure(figsize=(8,5))
cuisine_rating.head(10).plot(kind="bar", color="green")
plt.title("Top 10 Highest Rated Cuisine Combinations")
plt.xlabel("Cuisine Combinations")
plt.ylabel("Average Rating")
plt.xticks(rotation=35, ha="right")
plt.show()
```

## Conclusion

1. Certain cuisine combinations receive significantly higher average ratings than others.

2. The highest-rated cuisine combinations include Italian, Deli; Hawaiian, Seafood; American, Sandwich, Tea; Continental, Indian; and several European cuisine combinations, each with an average rating of 4.9.

3. Premium and international cuisine combinations generally tend to receive higher customer ratings.

4. This analysis indicates that cuisine combinations can influence customer satisfaction and restaurant ratings.

# Level_2 - Task_4: Restaurant Chains

## Objective

The objective of this task is to identify restaurant chains in the dataset and analyze their popularity and average ratings based on the number of restaurant outlets.

### 1. Identify restaurant chains

```python
restaurant_counts = df["Restaurant Name"].value_counts()
restaurant_chains = restaurant_counts[restaurant_counts > 1]
restaurant_chains.head(10)
```

### 2. Top 10 restaurant chains

```python
print("Top 10 Restaurant Chains")
restaurant_chains.head(10)
```

### 3. Visualize Top Restaurant Chains

```python
plt.figure(figsize=(8,5))
restaurant_chains.head(10).plot(kind="bar", color="orange")
plt.title("Top 10 Restaurant Chains")
plt.xlabel("Restaurant Chain")
plt.ylabel("Number of Outlets")
plt.xticks(rotation=45, ha="right")
plt.show()
```

### 4. Analyze Ratings and Popularity

```python
chain_analysis = df.groupby("Restaurant Name").agg({
    "Aggregate rating":"mean",
    "Votes":"sum"
})

chain_analysis = chain_analysis.loc[restaurant_chains.index]

chain_analysis = chain_analysis.sort_values(
    by=["Aggregate rating","Votes"],
    ascending=False
)

chain_analysis.head(10)
```

### 5. Visualize Highest Rated Restaurant Chains

```python
plt.figure(figsize=(8,5))
chain_analysis.head(10)["Aggregate rating"].plot( kind="bar", color="purple")
plt.title("Top Rated Restaurant Chains")
plt.xlabel("Restaurant Chain")
plt.ylabel("Average Rating")
plt.xticks(rotation=45, ha="right")
plt.show()
```

## Conclusion

The analysis identified several restaurant chains operating multiple outlets in the dataset. Cafe Coffee Day, Domino's Pizza, and Subway have the highest number of outlets, indicating their strong market presence. The rating analysis shows that restaurant chains such as Talaga Sampireun, AB's Absolute Barbecues, and Silantro Fil-Mex have the highest average ratings. Overall, while some chains are highly popular due to their large number of outlets, others stand out by maintaining consistently high customer ratings, reflecting strong customer satisfaction and service quality.
